**FUNCIÓN**

```python
def describe_dataframe(df):
    """
    Genera un resumen del DataFrame y muestra para cada columna:
    - Los tipos de datos
    - El porcentaje de valores nulos
    - La cardinalidad  
    - El número de valores únicos

    Argumentos:
    df (pandas.DataFrame): DataFrame que vamos a resumir

    Retorna:
    pandas.DataFrame: DataFrame con una columna por cada variable
    del DataFrame original. Como filas devuelve (definido en la función):
    - DATA_TYPE
    - MISSING(%)
    - CARDIN (%)
    - UNIQUE_VALUES
    """
    
    resumen = pd.DataFrame({
        "DATA_TYPE": df.dtypes,
        "MISSING(%)": df.isnull().mean() * 100,
        "CARDIN (%)": df.nunique() / len(df),
        "UNIQUE_VALUES": df.nunique()
    }).T

    return resumen
```
